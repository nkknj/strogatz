```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
```

```python
#4.1.2
def f(theta):
    return 1+2*np.cos(theta)

#circle
theta=np.arange(0, 2*pi, pi/180)
plt.plot(np.cos(theta), np.sin(theta), linestyle='-', color='black')

#point
theta_d=f(theta)
i=0
temps=[]
while i+1<theta.shape[0]:
    if theta_d[i]*theta_d[i+1]<=0:
        temp=theta[i]+(theta[i+1]-theta[i])*theta_d[i]/(theta_d[i]-theta_d[i+1])
        if theta_d[i+1]>0:
            plt.plot(np.cos(temp), np.sin(temp), marker='o', color='blue')
        else:
            plt.plot(np.cos(temp), np.sin(temp), marker='o', color='red')
        
        temps.append(temp)
    
    i+=1

#triangle
if len(temps)==0:
    if f(pi/2)>=0:
        plt.plot(np.cos(pi/2), np.sin(pi/2), marker=(3, 0, (pi/2)*180/pi), color='dodgerblue', markersize=10)
    else:
        plt.plot(np.cos(pi/2), np.sin(pi/2), marker=(3, 0, 180+(pi/2)*180/pi), color='dodgerblue', markersize=10)
elif len(temps)==1:
    if f(2*pi-temps)>=0:
        plt.plot(np.cos(2*pi-temps), np.sin(2*pi-temps), marker=(3, 0, (2*pi-temps)*180/pi), color='dodgerblue', markersize=10)
    else:
        plt.plot(np.cos(2*pi-temps), np.sin(2*pi-temps), marker=(3, 0, 180+(2*pi-temps)*180/pi), color='dodgerblue', markersize=10)
else:
    i=0
    while i+1<len(temps):
        midpoint=(temps[i+1]+temps[i])/2
        if f(midpoint)>=0:
            plt.plot(np.cos(midpoint), np.sin(midpoint), marker=(3, 0, midpoint*180/pi), color='dodgerblue', markersize=10)
        else:
            plt.plot(np.cos(midpoint), np.sin(midpoint), marker=(3, 0, 180+midpoint*180/pi), color='dodgerblue', markersize=10)
        
        i+=1
    
    midpoint=temps[0]/2+temps[-1]/2+pi
    if f(midpoint)>=0:
        plt.plot(np.cos(midpoint), np.sin(midpoint), marker=(3, 0, midpoint*180/pi), color='dodgerblue', markersize=10)
    else:
        plt.plot(np.cos(midpoint), np.sin(midpoint), marker=(3, 0, 180+midpoint*180/pi), color='dodgerblue', markersize=10)
        

plt.grid()
plt.gca().set_aspect('equal')
plt.show()
```
