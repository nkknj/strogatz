```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
inf=np.inf
```

```python
#3.1 サドルノード分岐
#3.1.1 ベクトル場は省略し分岐図のみ

def f(x, r):
    return 1+r*x+x**2

R=np.arange(-5, 5, 0.01)

for r in R:
    x=np.arange(-5, 5, 0.1)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=0:
            tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x*')
plt.grid()
plt.show()
```

```python
#3.4.11(a)

def f(x, r):
    return r*x-np.sin(x)

r=0
x=np.arange(-10, 10, 0.1)
x_d=f(x, r)
plt.plot(x, x_d, linestyle='-', color='b', label='x_d')

i=0
fixpoint=[]
while i+1<x.shape[0]:
    if x_d[i]*x_d[i+1]<=0:
        fixpoint.append(x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1]))
        
    i+=1

fixpoint=np.array(fixpoint)

if fixpoint.shape[0]==0:
    xcenter=int(x.shape[0]/2)
    if f(x[xcenter], r)>0:
        plt.plot(xcenter, 0, marker='>', color='b')
    elif f(x[xcenter], r)<0:
        plt.plot(xcenter, 0, marker='<', color='b')
    else:
        pass
elif fixpoint.shape[0]==1:
    leftcenter=(fixpoint-x[0])/2
    if f(leftcenter, r)>0:
        plt.plot(leftcenter, 0, marker='>', color='b')
    elif f(leftcenter, r)<0:
        plt.plot(leftcenter, 0, marker='<', color='b')
    else:
        pass
else:
    #left side vector marker
    position=fixpoint[0]-(fixpoint[1]-fixpoint[0])/2
    if f(position, r)>0:
        plt.plot(position, 0, marker='>', color='b')
    else:
        plt.plot(position, 0, marker='<', color='b')
    
    #markers between fixpoints
    i=0
    while i+1<fixpoint.shape[0]:
        midpoint=(fixpoint[i]+fixpoint[i+1])/2
        if f(midpoint, r)>0:
            plt.plot(midpoint, 0, marker='>', color='b')
        else:
            plt.plot(midpoint, 0, marker='<', color='b')
        i+=1
    
    #right side vector marker
    position=fixpoint[-1]+(fixpoint[-1]-fixpoint[-2])/2
    if f(position, r)>0:
        plt.plot(position, 0, marker='>', color='b')
    else:
        plt.plot(position, 0, marker='<', color='b')
    
plt.legend()
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
#3.4.12
#n分岐

def f(x, r, n):
    k=n%2
    n=int(n/2)
    temp=1
    
    for i in range(n):
        temp*=((i+1)*r-x**2)

    return temp*(x**k)

R=np.arange(-2, 2, 0.01)
x=np.arange(-2, 2, 0.01)
N=np.arange(5, 20, 1)

for n in N:
    for r in R:
        x_d=f(x, r, n)

        i=1
        while i+1<x.shape[0]:
            if x_d[i]*x_d[i+1]<=0:
                tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
                if x_d[i]>x_d[i+1]:
                    plt.plot(r, tempx, marker='o', color='red', markersize=2)
                else:
                    plt.plot(r, tempx, marker='o', color='blue', markersize=2)
            i+=1

    plt.xlabel('r')
    plt.ylabel('x*')
    plt.title(str(n)+'-furcation')
    plt.grid()
    plt.show()
```

```python
#3.4.13(a)

def f(x, r):
    return r-x-np.exp(-1*x)

def fd(x, r):
    return -1+np.exp(-1*x)

#for Newton-Raphson method
#In: position (x, y) and grad
#Out: x position where x axis crosses the line made by In position and grad
def NR(x, y, grad):
    return x-y/grad

R=np.arange(0.01, 3, 0.01)
NRtimes=10

for r in R:
    x=np.arange(-2, 2, 0.01)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=0:
            tempx=x[i]
            for t in range(NRtimes):
                y=f(tempx, r)
                grad=fd(tempx, r)
                tempx=NR(tempx, y, grad)
                
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x*')
plt.grid()
plt.show()
```
