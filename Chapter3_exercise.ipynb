```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
inf=np.inf
```

```python
#3.1 サドルノード分岐
#3.1.1
def f(x, r):
    return 1+r*x+x**2

R=np.arange(-5, 5, 0.01)

for r in R:
    x=np.arange(-5, 5, 0.1)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=0:
            tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x')
plt.grid()
plt.show()
```

```python
#3.1.5 (b)
#0になってくれないので1e-4以下はゼロとして扱うことで回避している

def f(x, r):
    return r**2+x**2

R=np.arange(-1, 1, 0.1)

for r in R:
    x=np.arange(-1, 1, 0.1)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=1e-5:
            tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x')
plt.grid()
plt.show()
```
