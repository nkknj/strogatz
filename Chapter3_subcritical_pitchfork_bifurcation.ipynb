```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#式(3,4,2)の分岐図
def f(x, r):
    return r*x+x**3

R=np.arange(-1, 1, 0.001)

for r in R:
    x=np.arange(-2, 2, 0.01)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=0:
            tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x*')
plt.grid()
plt.show()
```

```python
#式(3,4,3)の分岐図
def f(x, r):
    return r*x+x**3-x**5

R=np.arange(-1, 1, 0.001)

for r in R:
    x=np.arange(-2, 2, 0.01)
    x_d=f(x, r)

    i=1
    while i+1<x.shape[0]:
        if x_d[i]*x_d[i+1]<=0:
            tempx=x[i]+(x[i+1]-x[i])*x_d[i]/(x_d[i]-x_d[i+1])
            if x_d[i]>x_d[i+1]:
                plt.plot(r, tempx, marker='o', color='red', markersize=2)
            else:
                plt.plot(r, tempx, marker='o', color='blue', markersize=2)
        i+=1

plt.xlabel('r')
plt.ylabel('x*')
plt.grid()
plt.show()
```
