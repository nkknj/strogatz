```python
#-*- coding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
```

```python
#2.1 幾何学的な考察法
#2.1.1 -> blue line, red circles and crosses
#2.1.2 -> magenta squares
#2.1.3 -> tan line

#ベクトル場を描出
x=np.arange(-2.5*pi, 2.5*pi, pi/10)
x_d=np.sin(x)
plt.plot(x, x_d, linestyle='-', color='b', label='x_d')

#x軸上に流れ方向へ三角形を描出
for i in range(x.shape[0]):
    if x_d[i]>1e-4:
        plt.plot(x[i], 0, marker='>', color='dodgerblue')
    elif x_d[i]<-1e-4:
        plt.plot(x[i], 0, marker='<', color='dodgerblue')
    else:
        #流れがゼロの点
        if x_d[i+1]>0:
            #直後が正の時
            plt.plot(x[i], 0, color='red', marker='x', markersize=5)
        else:
            #直後が負の時
            plt.plot(x[i], 0, color='red', marker='o', markersize=5)

#2.1.2
#x_dが最大の時True, それ以外はFalseとなるラベルを作成
x_d_maxlabel=(x_d>np.max(x_d)-(1e-4))

#x_d_maxlabelがTrueの点に四角を描出
plt.plot(x[x_d_maxlabel], x_d[x_d_maxlabel], color='magenta', linestyle='None', marker='s', markersize=5)

plt.legend()
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
#2.1.3
#ベクトル場を描出
x=np.arange(-2.5*pi, 2.5*pi, pi/20)
x_d=np.sin(x)
plt.plot(x, x_d, linestyle='-', color='b', label='x_d')
x_dd=np.cos(x)*np.sin(x)
plt.plot(x, x_dd, linestyle='-', color='tan', label='x_dd')

#x_ddが最大の時True, それ以外はFalseとなるラベルを作成
x_dd_maxlabel=(x_dd>np.max(x_dd)-(1e-4))

#x_dd_maxlabelがTrueの点に四角を描出
plt.plot(x[x_dd_maxlabel], x_dd[x_dd_maxlabel], marker='s', linestyle='None', color='magenta', markersize=5)

plt.legend()
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
#2.2 固定点と安定性
#2.2.1
x=np.arange(-4, 4, 0.2)
x_d=4*x**2-16
plt.plot(x, x_d, linestyle='-', color='b', label='x_d')

#x軸上に流れ方向へ三角形を描出
for i in range(x.shape[0]):
    if x_d[i]>0:
        plt.plot(x[i], 0, marker='>', color='dodgerblue')
    elif x_d[i]<0:
        plt.plot(x[i], 0, marker='<', color='dodgerblue')
    else:
        #流れがゼロの点
        if x_d[i+1]>0:
            #直後が正の時
            plt.plot(x[i], 0, color='red', marker='s', markersize=5)
        else:
            #直後が負の時
            plt.plot(x[i], 0, color='red', marker='o', markersize=5)

i=0
while i+1<x.shape[0]:
    if x_d[i]*x_d[i+1]<0:
        if x_d[i]>0:
            plt.plot(x[i]+(x[i+1]-x[i])*(x_d[i]/(x_d[i]-x_d[i+1])), 0, color='red', marker='o', markersize=5)
        else:
            plt.plot(x[i]+(x[i+1]-x[i])*(-x_d[i]/(x_d[i+1]-x_d[i])), 0, color='red', marker='s', markersize=5)
    
    i+=1
    
plt.legend()
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
#2.8 コンピューターによる方程式の解法
#2.8.2(a)

l=0.2

t=np.arange(-5, 5, 0.5)
x=np.arange(-3, 3, 0.5)
x_d=x

for i in range(t.shape[0]):
    for j in range(x.shape[0]):
        theta=np.arctan(x_d[j])
        point_x=np.array([t[i]-l*np.cos(theta), t[i]+l*np.cos(theta)])
        point_y=np.array([x[j]-l*np.sin(theta), x[j]+l*np.sin(theta)])
        plt.plot(point_x, point_y, linestyle='-', color='green')

plt.show()
```
