```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#6.1.1
def f(x, y):
    return x-y

def g(x, y):
    return 1-np.exp(x)

def dfdx(x, y):
    delta=1e-4
    return (f(x+delta, y)-f(x-delta, y))/(2*delta)

def dfdy(x, y):
    delta=1e-4
    return (f(x, y+delta)-f(x, y-delta))/(2*delta)

def dgdx(x, y):
    delta=1e-4
    return (g(x+delta, y)-g(x-delta, y))/(2*delta)

def dgdy(x, y):
    delta=1e-4
    return (g(x, y+delta)-g(x, y-delta))/(2*delta)

X=np.arange(-2, 2, 0.1)
Y=np.arange(-2, 2, 0.1)
X, Y=np.meshgrid(X, Y)

#
plt.quiver(X, Y, f(X, Y), g(X, Y))

#
plt.contour(X, Y, f(X, Y), levels=[0], colors='green')
plt.contour(X, Y, g(X, Y), levels=[0], colors='orange')

#(x*, y*)=(0, 0)
fixpoint=np.array([0, 0])
J=np.array([
    [dfdx(fixpoint[0], fixpoint[1]), dfdy(fixpoint[0], fixpoint[1])],
    [dgdx(fixpoint[0], fixpoint[1]), dgdy(fixpoint[0], fixpoint[1])]
])

tau=J[0, 0]+J[1, 1]
D=np.linalg.det(J)

if D>0:
    if tau==0:
        print(str(fixpoint)+': center')
        plt.plot(fixpoint[0], fixpoint[1], marker='o', color='green')
    elif tau>0:
        if tau**2-4*D>0:
            print(str(fixpoint)+': unstable node')
            plt.plot(fixpoint[0], fixpoint[1], marker='o', color='blue')
        else:
            print(str(fixpoint)+': unstable spiral')
            plt.plot(fixpoint[0], fixpoint[1], marker='s', color='blue')
    else:
        #tau<0
        if tau**2-4*D<0:
            print(str(fixpoint)+': stable node')
            plt.plot(fixpoint[0], fixpoint[1], marker='o', color='red')
        else:
            print(str(fixpoint)+': stable spiral')
            plt.plot(fixpoint[0], fixpoint[1], marker='s', color='red')
elif D<0:
    print(str(fixpoint)+': saddle')
    plt.plot(fixpoint[0], fixpoint[1], marker='o', color='magenta')
else:
    print('Check')
    plt.plot(fixpoint[0], fixpoint[1], marker='*', color='magenta')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('vector field')
plt.grid()
plt.show()
```
