```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#5.1.7
A=np.array([
    [1, 0],
    [1, 1]
])

X=np.linspace(-1, 1, 9)
Y=np.linspace(-1, 1, 9)

#ベクトル場の描出
for x in X:
    for y in Y:
        grad=np.dot(A, np.array([x, y]))
        plt.quiver(x, y, grad[0], grad[1], angles='xy', scale_units='xy', scale=5)
        
#典型的な軌道
a=np.array([0.2, 0.01]).reshape(2, 1)
for t in range(100):
    temp=np.dot(A, a[:, -1])
    temp=a[:, -1]+temp*0.01
    a=np.c_[a, temp]
    
plt.plot(a[0], a[1], linestyle='-', color='magenta')

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid()
plt.show()
```
