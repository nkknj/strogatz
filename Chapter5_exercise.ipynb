```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#5.1.7
A=np.array([
    [1, 0],
    [1, 1]
])

X=np.linspace(-1, 1, 9)
Y=np.linspace(-1, 1, 9)

#ベクトル場の描出
for x in X:
    for y in Y:
        grad=np.dot(A, np.array([x, y]))
        plt.quiver(x, y, grad[0], grad[1], angles='xy', scale_units='xy', scale=5)
        
#典型的な軌道
a=np.array([0.2, 0.01]).reshape(2, 1)
for t in range(100):
    temp=np.dot(A, a[:, -1])
    temp=a[:, -1]+temp*0.01
    a=np.c_[a, temp]
    
plt.plot(a[0], a[1], linestyle='-', color='magenta')

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid()
plt.show()
```

```python
#5.2.14
#一様分布[-1, 1]

N=10000000
n_saddle=0
n_stable_node=0
n_stable_spiral=0
n_unstable_node=0
n_unstable_spiral=0

for i in range(N):
    A=np.random.rand(4).reshape(2, 2)*2-1
    t=A.trace()
    d=np.linalg.det(A)
    if d<=0:
        n_saddle+=1
    else:
        if t<=0:
            if t**2-4*d>0:
                n_stable_node+=1
            else:
                n_stable_spiral+=1
        else:
            if t**2-4*d>0:
                n_unstable_node+=1
            else:
                n_unstable_spiral+=1

print('n_saddle='+str(n_saddle/N))
print('n_stable_node='+str(n_stable_node/N))
print('n_stable_spiral='+str(n_stable_spiral/N))
print('n_unstable_node='+str(n_unstable_node/N))
print('n_unstable_spiral='+str(n_unstable_spiral/N))
```

```python
#5.2.14
#正規分布(0, 1)

N=10000000
n_saddle=0
n_stable_node=0
n_stable_spiral=0
n_unstable_node=0
n_unstable_spiral=0

for i in range(N):
    A=np.random.randn(4).reshape(2, 2)
    t=A.trace()
    d=np.linalg.det(A)
    if d<=0:
        n_saddle+=1
    else:
        if t<=0:
            if t**2-4*d>0:
                n_stable_node+=1
            else:
                n_stable_spiral+=1
        else:
            if t**2-4*d>0:
                n_unstable_node+=1
            else:
                n_unstable_spiral+=1

print('n_saddle='+str(n_saddle/N))
print('n_stable_node='+str(n_stable_node/N))
print('n_stable_spiral='+str(n_stable_spiral/N))
print('n_unstable_node='+str(n_unstable_node/N))
print('n_unstable_spiral='+str(n_unstable_spiral/N))
```

```python
#5.3.2
#不安定スパイラル

A=np.array([
    [0, 1],
    [-1, 1]
])

RJ=np.array([1, 0]).reshape(2, 1) # (R, J).T の初期条件
dt=0.001
Tmax=20
T=np.arange(0, Tmax, dt)

for t in T:
    x=RJ[:, -1]
    
    #Runge-Kutta method
    k1=np.dot(A, x)
    temp=x+k1*dt/2
    k2=np.dot(A, temp)
    temp=x+k2*dt/2
    k3=np.dot(A, temp)
    temp=x+k3*dt
    k4=np.dot(A, temp)
    x=x+(k1+2*k2+2*k3+k4)*dt/6
    
    RJ=np.c_[RJ, x.reshape(2, 1)]
    t+=dt

plt.plot(np.arange(0, Tmax+dt, dt), RJ[0], color='b', label='Romeo')
plt.plot(np.arange(0, Tmax+dt, dt), RJ[1], color='r', label='Juliet')
plt.legend()
plt.title('Romeo and Juliet')
plt.show()
```
